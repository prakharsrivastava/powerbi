## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
# File location and type
file_location = "/FileStore/tables/Automobile_data__1_.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
df=df.withColumnRenamed("fuel-type","fuel")
df=df.withColumnRenamed("num-of-doors","door")
```

```
# Create a view or table

temp_table_name = "Automobile_data__1__csv"

df.createOrReplaceTempView(temp_table_name)
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already loaded the data from the CSV file into a Pandas DataFrame 'df2'.

# Convert 'horsepower' column to numeric
df2['horsepower'] = pd.to_numeric(df2['horsepower'], errors='coerce')

# Drop rows with missing data in 'horsepower' column
df2.dropna(subset=['horsepower'], inplace=True)
df2.dropna(subset=['body-style'], inplace=True)

# Check the data type of 'horsepower' column
print(df2['horsepower'].dtype)

# Create a scatter plot
sns.boxplot(x='num-of-doors', y='horsepower', data=df2,hue='fuel')

```

```
%sql
SELECT 
    door,
    fuel,
    PERCENTILE_CONT(0.25) WITHIN GROUP (ORDER BY horsepower) AS q1,
    PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY horsepower) AS median,
    PERCENTILE_CONT(0.75) WITHIN GROUP (ORDER BY horsepower) AS q3,
    PERCENTILE_CONT(0.25) WITHIN GROUP (ORDER BY horsepower) - 1.5 * (PERCENTILE_CONT(0.75) WITHIN GROUP (ORDER BY horsepower) - PERCENTILE_CONT(0.25) WITHIN GROUP (ORDER BY horsepower)) AS lower_whisker,
    PERCENTILE_CONT(0.75) WITHIN GROUP (ORDER BY horsepower) + 1.5 * (PERCENTILE_CONT(0.75) WITHIN GROUP (ORDER BY horsepower) - PERCENTILE_CONT(0.25) WITHIN GROUP (ORDER BY horsepower)) AS upper_whisker
FROM
    Automobile_data__1__csv
WHERE
    horsepower IS NOT NULL and  door !="?" 
GROUP BY
    door,
    fuel
```

```
%sql
SELECT
make,
  gas AS gas,
  diesel AS diesel
FROM
  (
    SELECT fuel, horsepower,make
    FROM `Automobile_data__1__csv`
  ) src
PIVOT
(
  SUM(horsepower)
  FOR fuel IN ('gas', 'diesel')
) 
```
